```python
df.columns = df.columns.str.strip()
df
```

```python
print(df.duplicated().sum())
```

```python
# Adding Day, Month, and Year
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year

# Importing Warnings library
import warnings

# Ignoring unwanted warnings
warnings.filterwarnings("ignore")

df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
mean_unemployment_over_time = df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()
plt.plot(mean_unemployment_over_time.index, mean_unemployment_over_time.values,marker='o' ,linestyle='-')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.grid(True)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='Region', y='Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate by State')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(df['Estimated Unemployment Rate (%)'], kde=True)
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.grid(True)
```

```python
sns.countplot(x='Area', data=df)
plt.title('Countplot of Area')
plt.xlabel('Area')
plt.ylabel('Count')
```
